```python
import sys 
import numpy as np # linear algebra
from scipy.stats import randint
import matplotlib.pyplot as plt # this is used for the plot the graph 
%matplotlib inline
import seaborn as sns
from scipy.stats import chi2
import tensorflow as tf
from scipy import stats
import os 
import re
```

```python
from scipy import interpolate
```

```python
cnc=np.load('cn/cnc.npy')
cnq=np.load('cn/cnq.npy')
cnl=np.load('cn/cnl.npy')
```

```python
enc=np.load('EN_cr_cn/enc.npy')
enq=np.load('EN_cr_cn/enq.npy')
enl=np.load('EN_cr_cn/enl.npy')
encrc=np.load('EN_cr_cn/encrc.npy')
encrq=np.load('EN_cr_cn/encrq.npy')
encrl=np.load('EN_cr_cn/encrl.npy')
encnc=np.load('EN_cr_cn/encnc.npy')
encnq=np.load('EN_cr_cn/encnq.npy')
encnl=np.load('EN_cr_cn/encnl.npy')
```

```python
dpc=np.load('dp_and_cr/dpc.npy')
dpq=np.load('dp_and_cr/dpq.npy')
dpl=np.load('dp_and_cr/dpl.npy')
dpcrc=np.load('dp_and_cr/dpcrc.npy')
dpcrq=np.load('dp_and_cr/dpcrq.npy')
dpcrl=np.load('dp_and_cr/dpcrl.npy')
```

```python
per=np.linspace(0.02,0.98,6)
```

```python
def cal(x):
    return np.abs(x-per).mean()
```

```python
def nt(x):
    return(interpolate.interp1d(per,x)(0.9))
```

```python
cal(cnc),cal(enc),cal(dpc),cal(dpcrc),cal(encrc),cal(encnc)
```

```python
nt(cnc),nt(enc),nt(dpc),nt(dpcrc),nt(encrc),nt(encnc)
```

```python


fig = plt.gcf()
fig.set_size_inches(7.2,7.2)
plt.rcParams['axes.labelsize'] = 18

plt.rcParams['axes.titlesize'] = 18

plt.rcParams['xtick.labelsize'] = 11

plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 15

#plt.title('t=1')
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

plt.subplot(1, 1, 1)

plt.plot(per,np.abs(per-cnc),'-o',label='CN',color='black',linewidth=4,markersize=12)
plt.plot(per,np.abs(per-dpc),'-X',label='DP',color='red',linewidth=4,markersize=12)
plt.plot(per,np.abs(per-enc),'-d',label='EN',color='slateblue',linewidth=4,markersize=12)
plt.plot(per,np.abs(per-encnc),'-.*',label='EN-CN',color='maroon',linewidth=4,markersize=12)
plt.plot(per,np.abs(per-encrc),'-.>',label='EN-CR',color='deeppink',linewidth=4,markersize=12)
plt.plot(per,np.abs(per-dpcrc),'-.D',label='DP-CR',color='darkorange',linewidth=4,markersize=12)





plt.plot([0,1],[0,0],'--',color='black',linewidth=0.8)

plt.xlabel('Nominal Coverage')
plt.ylabel('| True Coverage - Nominal Coverage |')
plt.legend(loc=2)
fig.savefig('mpgc.pdf',transparent=True)
```
